# 9.2 

Define a generator function called get_odds() that returns the odd numbers from range(10). 
Use a for loop to find and print the third value returned.

```python
def get_odds():
    for num in range(10):
        if num % 2 == 1:
            yield num

count = 0
for value in get_odds():
    count += 1
    if count == 3:
        print(value)
        break
```
